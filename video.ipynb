```python
import shutil
import os
import ffmpeg

results = os.path.join('results')

def create_video(source):
    images = os.path.join(results, f'result_{source}', 'video')
    output = os.path.join(results, f'{source}.mp4')
    
    ffmpeg.input(os.path.join(images, '*.png'), pattern_type='glob', framerate=2) \
        .output(output, pix_fmt='yuv420p', vcodec='libx264').run()
    
    
def fix_filenames(root):
    for entry in os.scandir(root):
        if entry.name.startswith("gen"):
            try:
                tokens = entry.name.split('.')
                name = tokens[0].split('gen')[1]
                ext = tokens[1]
                if name.isdigit():
                    name = f'gen{name.zfill(6)}.{ext}'
                    dirname = os.path.dirname(entry.path)
                    rename = os.path.join(dirname, name)
                    os.rename(entry.path, rename)

            except:
                pass
            
def fix_source(source):
    root = os.path.join(results, f'result_{source}')
    vidims = os.path.join(root, 'video') 
    
    fix_filenames(root)
    fix_filenames(vidims)
```
